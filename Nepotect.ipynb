```
#Redução das colunas para melhorar o processamento

data = pd.read_csv("Congo.csv").drop(
    columns=["Município", "CPF", "Matrícula", "Unidade Orçamentária", 
             "Data de admissão", "Unidade Gestora"])
```

```
juncoes = set(["de", "do", "da", "dos", "das"])


def conta_sobrenomes(nome):
  nome = nome.split(" ")
  nome = set(nome) - juncoes
  for n in nome:
    ocorrencias[n] += 1
```

```
ocorrencias = defaultdict(int)
for servidor in data["Servidor"]:
  conta_sobrenomes(servidor)
```

```
freq = pd.DataFrame([i for i in ocorrencias.items()], columns=["Nome", "Ocorrências"])
freq = freq.set_index("Nome")
freq[freq['Ocorrências'] > 20 ].plot(kind="barh", figsize=(18,14))
```
